```python
# Найдем выборы в столбце для значений пульса
q1 = horse_data['pulse'].quantile(0.25)
q3 = horse_data['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse_data[horse_data['pulse'].between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
remove_outliers
```

```python
# Выброс для пульса есть, но тем не менее, если верить интернету, такие большие значения все еще реалистичные 
pd.concat([horse_data, remove_outliers]).drop_duplicates(keep=False)
```

```python
# Найдем выборы в столбце для значений температуры
q1 = horse_data['respiratory rate'].quantile(0.25)
q3 = horse_data['respiratory rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse_data[horse_data['respiratory rate'].between(lower_bound, upper_bound, inclusive=True)].sort_values('respiratory rate')
remove_outliers
```

```python
# Эти значения не похожи на выбросы, может это просто такой вид распределения?
pd.concat([horse_data, remove_outliers]).drop_duplicates(keep=False).sort_values('respiratory rate').head(20)
```

```python
# Найдем выборы в столбце для значений возраста
q1 = horse_data['Age'].quantile(0.25)
q3 = horse_data['Age'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse_data[horse_data['Age'].between(lower_bound, upper_bound, inclusive=True)].sort_values('Age')
remove_outliers
```

```python
pd.concat([horse_data, remove_outliers]).drop_duplicates(keep=False).sort_values('Age')
```

```python
# Посчитаем матрицу корреляций, чтобы понять, как данные столбцов влияют друг на друга.
# При этом, будем помнить, что столбцы 'surgery', 'Age', 'outcome' практически не имеют пустых значений
# и их можно использовать в качестве основных данных для подсчета корреляции
```

```python
corr_matr = horse_data.corr()
corr_matr
```

```python
# Таблица показывает, что данные столбца 'rectal temperature' практически никак не связаны с остальными данными,
# к тому же 'rectal temperature' имеет небольшой разброс значений, поэтому отсутствующие значения можно заменить
# медианой
```

```python
horse_data_filled = pd.read_csv('../../horse_data.csv', usecols=[0, 1, 3, 4, 5, 6, 10, 22], header=None, na_values='?')
horse_data_filled.columns = ['surgery', 'Age', 'rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain', 'outcome']
horse_data_filled['rectal temperature'].fillna(horse_data['rectal temperature'].median(), inplace=True)
```

```python
horse_data_filled['surgery'].fillna(horse_data['surgery'].mode()[0], inplace=True) # Заменим модой столбца
horse_data_filled['outcome'].fillna(horse_data['outcome'].mode()[0], inplace=True) # Заменим модой столбца
```

```python
horse_data['rectal temperature'].describe()
```

```python
horse_data_filled['rectal temperature'].describe()
```

```python
# Таким образом, мы видим, что при замене пустых значений медианой, итоговые характеристики столбца 'rectal temperature'
# изменились незначительно
```

```python
# Для столбца 'pulse' мы видим из таблицы коррелций, что его значения относительно хорошо связаны с даными 'Age'.
# Поэтому попробуем провести группировку по возрасту и определить недостающие данные медианой 'pulse'
```

```python
print(horse_data.groupby('Age')['pulse'].median())
```

```python
horse_data_filled['pulse'].fillna(horse_data.groupby('Age')['pulse'].transform('median'), inplace=True)
```

```python
horse_data['pulse'].describe()
```

```python
horse_data_filled['pulse'].describe()
```

```python
# Для столбца 'respiratory rate' таблица корреляции дает наибольшее сходство с данными столбца 'pulse', однако,
# с учетом того, что данные 'pulse' неполные и сами заполняются приближенными значениями,
# используем следующую по величине корреляцию с данными столбца 'Age'
```

```python
print(horse_data.groupby('Age')['respiratory rate'].median())
```

```python
horse_data_filled['respiratory rate'].fillna(horse_data.groupby('Age')['respiratory rate'].transform('median'), inplace=True)
```

```python
horse_data['respiratory rate'].describe()
```

```python
horse_data_filled['respiratory rate'].describe()
```

```python
# Для столбца 'temperature of extremities' таблица корреляции дает наибольшее сходство с данными столбца 'outcome',
# Поэтому попробуем провести группировку по исходу заболевания и доопределить данные медианой значения 'outcome'
```

```python
print(horse_data.groupby('outcome')['temperature of extremities'].median())
```

```python
horse_data_filled['temperature of extremities'].fillna(horse_data.groupby('outcome')['temperature of extremities'].transform('median'), inplace=True)
```
