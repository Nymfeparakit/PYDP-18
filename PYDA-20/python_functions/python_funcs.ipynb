```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def getDocumentOwner(documents):
  docNum = input('Введите номер документа: ')

  for doc in documents:
    if doc['number'] == docNum:
      return f"Владелец документа: {doc['name']}"
    
  return 'Документ не найден в базе'


def getDocumentShelf(documents):
  docNum = input('Введите номер документа: ')

  for key, value in directories.items():
    if docNum in value:
      return f"Документ хранится на полке: {key}"
    
  return 'Документ не найден в базе'


def getDocumentsInfo(documents, directories):
  documentsListStr = ''
  for doc in documents:
    docNum = doc['number']
    shelf = 0
    for key, value in directories.items():
      if docNum in value:
        shelf = key
    if shelf == 0:
      documentsListStr += f"№: {docNum}, тип: {doc['type']}, владелец: {doc['name']}\n"
    else:
      documentsListStr += f"№: {docNum}, тип: {doc['type']}, владелец: {doc['name']}, полка хранения: {shelf}\n"
  return documentsListStr


def addNewShelf(directories):
  shelfNum = input('Введите номер полки: ')
  
  if shelfNum not in directories:
    directories[shelfNum] = []
    return f"Полка добавлена. Текущий перечень полок: {list(directories.keys())}"
  else:
    return f"Такая полка уже существует. Текущий перечень полок: {list(directories.keys())}"


def deleteShelf(directories):
  shelfNum = input('Введите номер полки: ')
  
  if shelfNum not in directories:
    return f"Такой полки не существует. Текущий перечень полок: {list(directories.keys())}"
  elif not directories[shelfNum]:
    directories.pop(shelfNum)
    return f"Полка удалена. Текущий перечень полок: {list(directories.keys())}"
  elif directories[shelfNum]:
    return f"На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {list(directories.keys())}"


def addNewDocument(documents, directories, docNum = 0, docType = 0, docOwner = 0, shelfNum = 0):
  if docNum == 0:
    docNum = input('Введите номер документа: ')
    docType = input('Введите тип документа: ')
    docOwner = input('Введите владельца документа: ')
    shelfNum = input('Введите полку для хранения: ')

  if shelfNum not in directories:
    return 'Такой полки не существует. Добавьте полку командой as.'
  else:
    doc = {'type': docType, 'number' : docNum, 'name' : docOwner}
    documents.append(doc)
    directories[shelfNum].append(docNum)
    return 'Текущий список документов:\n' + getDocumentsInfo(documents, directories)


def deleteDocument(documents, directories, docNum = 0):
  if docNum == 0:
    docNum = input('Введите номер документа: ')

  for doc in documents:
    if doc['number'] == docNum:
      for key, value in directories.items():
        if docNum in value:
          directories[key].remove(docNum)
      documents.remove(doc)
      return 'Документ удален.\n' + getDocumentsInfo(documents, directories) 
      
  return 'Документ не найден в базе. Текущий перечень документов:\n' + getDocumentsInfo(documents, directories)


def moveDocument(documents, directories):
  docNum = input('Введите номер документа: ')
  shelfNum = input('Введите номер новой полки: ')

  haveShelf = True
  haveDoc = False
  if shelfNum not in directories:
    haveShelf = False
  for key, value in directories.items():
    if docNum in value:
      haveDoc = True

  if not haveShelf:
    return f"Такой полки не существует.\nТекущий перечень полок: {list(directories.keys())}\n" + getDocumentsInfo(documents, directories) 
  elif not haveDoc:
    return 'Документ не найден в базе.\n' + getDocumentsInfo(documents, directories)
  elif haveDoc and haveShelf:
    for doc in documents:
      if doc['number'] == docNum:
        docType = doc['type']
        docOwner = doc['name']
    deleteDocument(documents, directories, docNum)
    addNewDocument(documents, directories, docNum, docType, docOwner, shelfNum)    
    return 'Документ перемещен.\n' + getDocumentsInfo(documents, directories)


documentCmdsDict = {
    'p': getDocumentOwner,
    's' : getDocumentShelf,
}

directoryCmdsDict = {
    'ads' : addNewShelf,
    'ds' : deleteShelf,
}

docDirCmdsDict = {
    'ad' : addNewDocument,
    'd' : deleteDocument,
    'm' : moveDocument,
    'l' : getDocumentsInfo,
}

command = ''
while command != 'q':
  command = input('Введите команду: ')
  if command in documentCmdsDict:
    print(documentCmdsDict[command](documents))
  elif command in directoryCmdsDict:
    print(directoryCmdsDict[command](directories))
  elif command in docDirCmdsDict:
    print(docDirCmdsDict[command](documents, directories))
  elif command == 'q':
    print('Выход из программы')
  else:
    print('Такой команды нет')
```
