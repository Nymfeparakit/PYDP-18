### Задание 1

```python
def get_rating_str(rating_num):
    rating_dict = {
        rating_num <= 2: "низкий рейтинг",
        2 < rating_num <= 4: "средний рейтинг",
        4 < rating_num: "высокий рейтинг" 
    }
    
    return rating_dict[True]
```

```python
get_rating_str(2)
```

```python
data = pd.read_csv('ratings.csv')
data.head()
```

```python
data['class'] = data['rating'].apply(get_rating_str)
```

### Задание 2

```python
geo = pd.read_csv('keywords.csv')
```

```python
def set_region_data(query):
    geo_data = {
        'Центр': ['москва', 'тула', 'ярославль'],

        'Северо-Запад': ['петербург', 'псков', 'мурманск'],

        'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    }
    
    for key, value in geo_data.items():
        for city in geo_data[key]:
            if city in query.lower():
                return key
    return 'undefined'
```

```python
geo['region'] = geo['keyword'].apply(set_region_data)
geo.head()
```

### Задание 3

```python
import re
```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
def get_movie_year(film_title):
    year = re.search(r'\((\d{4})\)$', film_title)
    if year:
        year = int(year.group(1))
        if 1950 <= year <= 2010:
            return year
    return 1900   
```

```python
movies['year'] = movies['title'].apply(get_movie_year)
movies
```

```python
result = pd.merge(movies, ratings[['movieId','rating']], on=['movieId'])
result
```

```python
movies_grouped = result.groupby('year')[['rating']].mean().sort_values('rating', ascending = False)
movies_grouped
```
