```python
URL = 'https://habr.com/ru/all/'
KEYWORDS = ['python','–ø–∞—Ä—Å–∏–Ω–≥', 'cs-cart', '—Å—Ç—Ä–æ–≥–æ']
```

```python
def get_habr_news(link):
    # –°–¥–µ–ª–∞–µ–º get –∑–∞–ø—Ä–æ—Å, —Å–æ–∑–¥–∞–¥–∏–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
    req = requests.get(URL)
    soup = BeautifulSoup(req.text, 'html.parser')
    posts = soup.find_all('article', class_='post')
    habr_news = pd.DataFrame()

    # –≤—ã—Ç–∞—â–∏–º –∏–∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∑–∞–ø—Ä–æ—Å–∞ –Ω—É–∂–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ, –∑–∞–Ω–µ—Å–µ–º –∏—Ö –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
    for article in posts:
        time_data = article.find('span', class_='post__time')
        header_data = article.find('a', class_='post__title_link')
        preview_data = article.find('div', class_='post__text')
        
        for key_w in KEYWORDS:
            if key_w in header_data.text.lower() or key_w in preview_data.text.lower():
                frame_dict = {'date': time_data.text, 'title': header_data.text, 'link': header_data.attrs.get('href')}
                habr_news = pd.concat([habr_news, pd.DataFrame([frame_dict])])
                break
        
    return habr_news.reset_index()
    
```

```python
get_habr_news(URL)
```
