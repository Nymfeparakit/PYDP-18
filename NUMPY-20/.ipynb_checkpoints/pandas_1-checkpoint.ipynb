### Задание 1

```python
import pandas as pd
```

```python
data = pd.read_csv('/home/sasha/projects/data_science_course/ml-latest/ratings.csv')
data.head()
```

```python
movies = pd.read_csv('/home/sasha/projects/data_science_course/ml-latest/movies.csv')
movies.head()
```

```python
movies_info = movies[['movieId', 'title']]
```

```python
rated_data = data.query("rating == 5.0").groupby('movieId')[['movieId']].count().rename(columns={"movieId": "count"}).sort_values('count', ascending=False)
```

```python
rated_data = rated_data.merge(movies_info,  on='movieId', how='left')
rated_data.at[0,'title']
```

### Задание 2

```python
power_data = pd.read_csv('/home/sasha/projects/data_science_course/10_pandas_basics/power.csv')
power_data.head()
```

```python
pribalt_data = power_data[(power_data['country'] == 'Latvia') | (power_data['country'] == 'Lithuania') | (power_data['country'] == 'Estonia')]
pribalt_data.head()
```

```python
pribalt_data = pribalt_data.query('(category == 4) | (category == 12) | (category == 21)') \
    .query('year >= 2005 & year <= 2010') \
    .query('quantity > 0')
```

```python
print(pribalt_data['quantity'].sum())
```
